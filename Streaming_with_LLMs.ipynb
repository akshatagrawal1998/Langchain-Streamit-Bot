```python
# Import the 'os' module for environment variables and 'getpass' for secure password input
import os, getpass

# Prompt the user to securely enter their OpenAI API Key, which will not be displayed on screen
# This key is then stored as an environment variable for use by the LangChain library
os.environ["OPENAI_API_KEY"] = getpass.getpass("Enter your OpenAI API Key: ")
```

```python
# Invoke method
# Using the method invoke() will only return the completion once all of it has been generated.

# Import the ChatOpenAI class from the langchain_openai library
from langchain_openai import ChatOpenAI

# Initialize the ChatOpenAI model. This will use the API key set in the environment.
model = ChatOpenAI()

# Call the invoke method with a prompt. This will wait for the entire response to be generated
# before returning it as a single AIMessage object.
model.invoke("tell me a story")
```

```python
# Demonstrate the invoke() method again with a different prompt.
# As before, this prints the full, complete response after it has been entirely generated.
print(model.invoke("What is the capital?"))
```

```python
# Demonstrate the stream() method.
# Unlike invoke(), stream() returns a generator that yields chunks of the response as they are generated.
# Printing the generator directly shows its object representation, not the streamed content.
print(model.stream("tell me a story"))
```

```python
# To see the actual streamed content, we need to iterate through the yielded chunks from the stream() method.
# The next code cell demonstrates this by iterating over an asynchronous stream.
```

```python
# Utilize the astream() method for asynchronous streaming.
# This is particularly useful in environments where non-blocking operations are preferred.
# We iterate through each chunk as it's received and append it to a list, then print its content.
chunks = []
# The 'async for' loop is used to iterate over the asynchronous generator returned by astream().
# 'end="|"' adds a separator between chunks, and 'flush=True' ensures immediate output.


async for chunk in model.astream("hello. tell me what is the difference between Machine learning and Deep Learning?"):
    chunks.append(chunk)
    print(chunk.content, end="|", flush=True)
```

```python
# Access and display a specific chunk from the list of streamed chunks.
# This shows the content and metadata of an individual `AIMessageChunk` object.
chunks[2]
```

```python
# Demonstrate synchronous streaming using a traditional 'for' loop.
# This iterates over the generator returned by model.stream() and prints each content chunk immediately.
# 'end=""' prevents a newline after each chunk, and 'flush=True' ensures real-time output.


for chunk in model.stream("Write me a song about Machine Learning"):
    print(chunk.content, end="", flush=True)
```
